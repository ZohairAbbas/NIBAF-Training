```python
import re

def remove_tags(text):
  pattern = re.compile('<.*?>') # saving this pattern
  return pattern.sub(r'',text)  # wherever this pattern matches replace the character with a whitespace
```

URL remover

when working with api data like whats app , facebook, twitter , youtube data sets
you will find a lot of URLS in the data set so we need to remove the url
types of URLs
http://, https:// , www. , https://www.

```python
url='https://www.linkedin.com/in/razi-moosa-99656b194/'
url_1='http://www.linkedin.com/in/razi-moosa-99656b194/'
url_2='www.linkedin.com/in/razi-moosa-99656b194/'
url_3=url_1+url_2+'Razi'   # join the first two urls and text and see
url_4=url_3+' Razi here print this'
```

```python
url_3
```

```python
def remove_url_prefix(text):
  pattern = re.compile(r'https?://\S+|www\.\S+') # saving this pattern
  return pattern.sub(r'',text)                   # wherever this pattern matches replace the character with a whitespace
```

```python
print(remove_url_prefix(url)
,remove_url_prefix(url_1)
,remove_url_prefix(url_2), remove_url_prefix(url_3), remove_url_prefix(url_4))
```

Removing punctuation in text preprocessing helps simplify the input for NLP models, as punctuation often adds no semantic value and can interfere with tokenization. By removing it, models focus on the core meaning of the text, improving efficiency and accuracy in tasks like sentiment analysis or language modeling.

```python
import string #import this module

string.punctuation
```

```python
#why is it necessary to remove punctuation?
#'Hello! how are you?'
#Hello , ! , how , are , you , ?
#more tokens than required (!,?)
#Hello!, how , are , you?
#Hello! not equal to Hello and you? not equal to you!
#this is not a hard and fast rule just best practice depending your project scope
#we might not remove punctuation
```

```python
exclude=string.punctuation

def remove_punctuation(text): #copy this function as it is
  return text.translate(str.maketrans('','',exclude))
```

```python
sample_text='hello?, how are you!'
remove_punctuation(sample_text)
```

```python
df['list of values'][42:58].apply(remove_punctuation)
```

```python
#chat word treatment
#converting abbreviation or slangs into words
#ROFL, LOL, ASAP, GN
#Rolling on Floor laughing, laugh out loud, as soon as possible, good night
# Provided abbreviations
abbreviations = """
AFAIK=As Far As I Know
AFK=Away From Keyboard
ASAP=As Soon As Possible
ATK=At The Keyboard
ATM=At The Moment
A3=Anytime, Anywhere, Anyplace
BAK=Back At Keyboard
BBL=Be Back Later
BBS=Be Back Soon
BFN=Bye For Now
B4N=Bye For Now
BRB=Be Right Back
BRT=Be Right There
BTW=By The Way
B4=Before
CU=See You
CUL8R=See You Later
CYA=See You
FAQ=Frequently Asked Questions
FC=Fingers Crossed
FWIW=For What It's Worth
FYI=For Your Information
GAL=Get A Life
GG=Good Game
GN=Good Night
GMTA=Great Minds Think Alike
GR8=Great!
G9=Genius
IC=I See
ICQ=I Seek you (also a chat program)
ILU=ILU: I Love You
IMHO=In My Honest/Humble Opinion
IMO=In My Opinion
IOW=In Other Words
IRL=In Real Life
KISS=Keep It Simple, Stupid
LDR=Long Distance Relationship
LMAO=Laugh My A.. Off
LOL=Laughing Out Loud
LTNS=Long Time No See
L8R=Later
MTE=My Thoughts Exactly
M8=Mate
NRN=No Reply Necessary
OIC=Oh I See
PITA=Pain In The A..
PRT=Party
PRW=Parents Are Watching
ROFL=Rolling On The Floor Laughing
ROTFLMAO=Rolling On The Floor Laughing My A.. Off
SK8=Skate
STATS=Your sex and age
ASL=Age, Sex, Location
THX=Thank You
TTFN=Ta-Ta For Now!
TTYL=Talk To You Later
U=You
U2=You Too
U4E=Yours For Ever
WB=Welcome Back
WTF=What The F...
WTG=Way To Go!
WUF=Where Are You From?
W8=Wait...
7K=Sick:-D Laugher
TFW=That feeling when
MFW=My face when
MRW=My reaction when
IFYP=I feel your pain
TNTL=Trying not to laugh
JK=Just kidding
IDC=I don‚Äôt care
ILY=I love you
IMU=I miss you
ADIH=Another day in hell
ZZZ=Sleeping, bored, tired
WYWH=Wish you were here
TIME=Tears in my eyes
BAE=Before anyone else
FIMH=Forever in my heart
BSAAW=Big smile and a wink
BWL=Bursting with laughter
BFF=Best friends forever
CSL=Can‚Äôt stop laughing
OMG=Oh my god
"""

# Split the lines and create a dictionary
abbreviation_dict = dict(line.split('=') for line in abbreviations.strip().split('\n'))

# Print the dictionary
print(abbreviation_dict)
```

```python
def chat_words(text):
  new_text=[]
  for w in text.split():
    if w.upper() in abbreviation_dict:
      new_text.append(abbreviation_dict[w.upper()])
    else:
        new_text.append(w)
  return " ".join(new_text)
```

Stop Words

```python
#removing stop words
#stop words are words that are not that useful in sentiment analysis
#stop are words like a, my, the, of, are
#in advanced nlp technique like part of speech tagging we dont remove
#as it helps in sentence formation but for our use case like document categorization
#or sentiment analysis we can safely remove it
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
stopwords=stopwords.words('english')
stopwords
```

```python
#remove stopwords
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords:
     pass
    else:
     new_text.append(word)
  temp=new_text[:]
  new_text.clear()
  return " ".join(temp)
```

```python
s1='hello, my name is razi and this is my car. my car is red.'
remove_stopwords(s1)
```

```python
df['list of values'][42:58].apply(remove_stopwords)
```

Handling emojis

```python
#handling emojis
# we can either remove emoji or replace emoji
```

```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u'\U00010000-\U0010ffff'
                           u"\u200d"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\u3030"
                           u"\ufe0f"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

Tokenization

```python
#tokenization is the most important part of text preprocessing
#Tokenization is a fundamental step in preparing text data for NLP tasks to make the text more suitable for machine learning models.
#sentences can be tokenized into word, phrases or individual sentence
#tokenization basically means to break a sentences into smaller segments
'Hello, I am studying in NIBAF. I love NIBAF'
['Hello','I', 'am', 'studying', 'in', 'NIBAF', 'Love'] # this is word tokenization unique words
[('Hello, I am studying in NIBAF'),('I love NIBAF')] # this is sentence tokenization
```

```python
#why do we have to do?
#Lets assume we want to make a text classification system which will make decisions like segregating customer queries into
#customer support or customer sales. We can figure this out based on tokenization.

'I am love karachi. I am in new karachi town.'
['I','am', 'love', 'karachi', 'in', 'new', 'town'] # based on unique word tokenization but we lose context as we can see that new karachi town
# is a place in  north eastern karachi. So this is why tokenization is needs to be done correctly otherwise your training model will get confused
# and will out put inaccurate results
```

```python
# challenges faced in tokenization
# prefix : $20 , NewYork , AbuDhabi
# suffix : 10 km , 100 miles, 2 hours
# Infix : U.S.A , Let's
```
